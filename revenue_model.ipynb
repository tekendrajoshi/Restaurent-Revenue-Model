```python
import pandas as pd
import numpy as np
df= pd.read_csv('data.csv')
df=df.drop('Name', axis=1)
df #ailesamma data ramrosanga bujhyo and unnecessary column i.e. name hatayo 
```

```python
# check for missing values
df.isnull().sum()  # no missing values in every column
```

```python
#aba ya vako location jasto column lai ni category wise numeric value dinuparo
#tesko lagi first check the categories
unique_locations = df['Location'].unique()
unique_locations
```

```python
unique_cuisine = df['Cuisine'].unique()
unique_cuisine
```

```python
#aba hamilai category haru thavo so encode them with number
from sklearn.preprocessing import LabelEncoder
encoder= LabelEncoder()
df['location_encoded']=encoder.fit_transform(df['Location'])
df['cuisine_encoded']=encoder.fit_transform(df['Cuisine'])
df['parking_encoded']=encoder.fit_transform(df['Parking Availability'])
df

```

```python
#aba edcode gareko column rakhne ra original column hataune
df=df.drop(['Location','Cuisine', 'Parking Availability'], axis=1)
df
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
df['revenue_scaled']= scaler.fit_transform(df[['Revenue']])
df
```

```python
#revenue svcaled vanne column banaisakyam so aba original wala drop garne
df=df.drop(['Revenue'], axis=1)
df
```

```python
import lightgbm as lgb
from sklearn.model_selection import train_test_split

# Split data
X = df.drop('revenue_scaled', axis=1)  # Features
y = df['revenue_scaled']               # Target
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

# Train LightGBM
model = lgb.LGBMRegressor(random_state=42)  # Default params work well!
model.fit(X_train, y_train)

# Predict
y_pred = model.predict(X_test)
```

```python
from sklearn.metrics import mean_absolute_error, r2_score
print("MAE:", mean_absolute_error(y_test, y_pred))
print("R² Score:", r2_score(y_test, y_pred))
```

```python
import joblib
joblib.dump(model, 'revenue_model.pkl')
```
